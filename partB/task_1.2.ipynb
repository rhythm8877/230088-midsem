## Assumption 1: Local Spatio-Temporal Patches Contain Sufficient Information for Action Discrimination

- **Assumption:** The ISA method assumes that small, local spatio-temporal patches (e.g., 16×16×T) extracted from video contain enough motion and appearance information to distinguish between different actions. The entire pipeline builds features bottom-up from these local patches.

- **Why the Method Needs It:** The ISA algorithm operates on fixed-size local patches — it learns filters by optimizing the ISA objective (Equation 1) over collections of these small patches. If the discriminative information for an action exists only in global context (e.g., the spatial relationship between a person's hand and a distant object), local patches would never capture it, and the learned features would be uninformative for classification.

- **Violation Scenario:** Consider an action recognition task where the difference between two actions lies entirely in the global spatial arrangement rather than local motion. For example, distinguishing "putting an object on a table" from "picking up an object from a table" — both have similar local hand motions, but the critical difference is the starting and ending global configuration. A dataset of such context-dependent actions would violate this assumption.

- **Paper Reference:** Section 2.1 and Section 3 describe the patch extraction process. The architecture in Figure 1 shows that the entire feature pipeline begins with local patch inputs. The assumption is implicit in the choice to use local receptive fields rather than processing entire frames.

## Assumption 2: Invariance Through Subspace Pooling Is Beneficial (Not Harmful) for the Target Task

- **Assumption:** The ISA method assumes that pooling features within learned subspaces (the square-root pooling in Eq. 1) to achieve invariance to local transformations (translation, phase shifts) is desirable for action recognition. In other words, it assumes that throwing away within-subspace variation improves — or at least does not harm — classification accuracy.

- **Why the Method Needs It:** The core novelty of ISA over ICA is precisely this pooling step. ISA groups features into subspaces and pools them, creating units that respond similarly to a set of related transformations (e.g., a gabor filter at different phases). If the variation that pooling discards is actually discriminative for the target task, then pooling would destroy critical information and degrade performance.

- **Violation Scenario:** Consider a task where the exact phase or precise spatial position of a texture is critical for classification — for instance, distinguishing between two very similar hand gestures that differ only in the precise finger positioning (a few pixels of translation). The ISA pooling would make the representation invariant to exactly the shift that distinguishes these gestures, leading to misclassification.

- **Paper Reference:** Equation (1) in Section 2.1, where the square root of the sum of squared activations within each subspace defines the ISA pooling. Section 2.2 discusses how ISA learns invariant features analogous to complex cells in V1, explicitly calling invariance a benefit.

## Assumption 3: Features Learned from Unlabeled Video Transfer to Supervised Action Classification

- **Assumption:** The ISA features are learned in a completely unsupervised manner — the algorithm never sees action labels during feature learning. The method assumes that the statistical structure captured by ISA from raw unlabeled video patches (i.e., independent subspace structure) aligns well with the visual features that are discriminative for supervised action recognition tasks.

- **Why the Method Needs It:** The entire ISA pipeline is unsupervised. The SVM classifier is trained on the learned features, but the features themselves are optimized for statistical independence, not for class discrimination. If the statistically independent subspaces do not correspond to action-relevant visual patterns, the features would be poor inputs for the downstream SVM, and the method would underperform compared to supervised or task-specific features.

- **Violation Scenario:** Consider a dataset where actions are distinguished not by low-level visual motion patterns but by semantic context (e.g., "cooking" vs. "chemistry experiment" — both involve similar hand-object interactions but differ in the type of objects and background context). The unsupervised ISA features, which optimize for generic statistical structure in motion patches, may not capture semantic or object-identity information needed for such fine-grained discrimination.

- **Paper Reference:** Section 1 (Introduction) states the features are learned in an unsupervised manner from unlabeled data. Section 4 (Experiments) shows the features trained on one set of data being used to classify actions, implicitly relying on the transferability assumption. The paper acknowledges that their features "discover useful groupings" but does not guarantee alignment with class boundaries.
